```python
#importing packages
import pandas as pd
import seaborn as sb
from matplotlib import pyplot as plt
#loading dataset using seaborn
df = sb.load_dataset('tips')


print(df.to_string())
#plotting how many time we got tip of n rs/-
g = sb.FacetGrid(df, col = 'time')
g.map(plt.hist, "tip")
plt.show()
```

```python
#gender based plot
g = sb.FacetGrid(df, col='sex', hue = "smoker")
g.map(plt.scatter, "total_bill","tip")
plt.show()
```

```python
g = sb.FacetGrid(df, col='time', hue = "size")
g.map(plt.scatter, "total_bill","tip")
plt.show()
```

```python
#STASTICAL ESTIMATION
sb.barplot(x = 'sex', y = "tip", hue = "time", data = df)
plt.show()
```

```python
#LINEAR RELATIONSHIP
sb.regplot( x = "total_bill", y ="tip",data =df)
plt.show()
```

```python
#pair plot
sb.pairplot(df)
plt.show()
```

```python
#pairplot seaborn:plotting selected variables
selected_vars = ['total_bill', 'tip']
sb.pairplot(df, vars=selected_vars)
plt.show()
```

```python
#adding a hue color to seaborn pairplot
sb.pairplot(df, hue="size")
plt.show()
```

```python
#plotting using different colors
sb.pairplot(df, hue = "size", palette="husl")
plt.show()
```

```python
#Diagonal kind of plots
sb.pairplot(df,diag_kind = "kde")
plt.show()
```

```python
#Adjusting plot kind
sb.pairplot(df, kind="reg")
plt.show()
```

```python
#controlling the markers
sb.pairplot(df,hue='sex', markers=["o", "s"])
plt.show()
```

```python
#advanced customization with seaborn pairplot
g = sb.pairplot(df, hue='day')
g.fig.suptitle("Pairplot of Tips Dataset", y=1.02)
g.set(xticks=[], yticks=[]) #remove tick labels
plt.show()
```
